**Exo 1.0 -- première sommation** Calculez la somme S des éléments du tableau avec une boucle. Vérifiez avec la fonction native `sum`.

```python
l = [3, 5, 3, -1, 6, 4, 9]
s = 0
for i in l:
    s += i

print(sum(l))

print(s)
```
